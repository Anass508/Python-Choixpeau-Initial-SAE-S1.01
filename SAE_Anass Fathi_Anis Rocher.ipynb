### Question 4

Définissez la fonction `maison` prenant en paramètre un tableau de réponses ainsi que l'indice d'un élève. La fonction retourne la maison correspondant à l'affectation de cet élève. Ainsi, l'appel de la fonction maison avec le tableau de réponses donné en exemple doit retourner : 
- `"Gryffondor"` avec l'indice `0` (correspondant à Harry Potter),
- `"Poufsouffle"` avec l'indice `5` (correspondant à Cedric Diggory),
- `"Serpentard"` avec l'indice `10` (correspondant à Drago Malefoy).

En effet, Harry Potter a mis le meilleur score à la question 1 et doit donc être affecté à Gryffondor. Cedric Diggory a mis le meilleur score à la question 3 et doit être affecté à la maison Poufsouffle. Finalement, Drago a mis le meilleur score à la question 4 et doit donc être affecté à la maison Serpentard.

Définissez également une fonction de tests unitaires pour la fonction `maison`.

````python
#Initialisation des fonctions : 

def maison(tab, indice_eleve):
    
    """
    Cette fonction donne la maison associée a un élève.
    tableau (entrée: 1) : tableau contenant le prénom et les notes des élèves , type list
    entrée 2: La position d'un élève dans notre tableau, type int
    valeur retournée : renvoie une chaine de caractère caractérisant la maison de l'élève, type str
    """

    nom_maison = ["Gryffondor", "Serdaigle", "Poufsouffle", "Serpentard"] #tableau des maisons de Poudlard
    ligne = tab[indice_eleve]                    
    tab_split1 = ligne.split(":") # séparation entre le nom et les notes de l'élève           
    tab_split2 = tab_split1[1].split("/") #séparation entre chaque note de l'élève
    max = int(tab_split2[0]) #initialisation de la plus grande note 
    i = 1
    pos = 0 #position de la meilleur note
    
    while i <= 3:
        if max < int(tab_split2[i]):
            max = int(tab_split2[i])
            pos = i
        i = i + 1
        
    return nom_maison[pos]

def test_maison():
    
    """
    Cette fonction teste le bon fonctionnement de la fonction maison()
    """
    
    assert maison(tab,0) == "Poufsouffle", "Erreur,la fonction ne renvoie pas les bons résultats (1)"
    assert maison(tab,3) == "Serpentard", "Erreur,la fonction ne renvoie pas les bons résultats (2)"
    print("La fonction lecture_reponses fonctionne correctement : \n ")

#Programme Principal : 

nom_fichier = "questionnaire_premiere_annee.txt" 
tab = lecture_reponses(nom_fichier)
test_maison() #lancement de la fonction de test pour maison()
indice_eleve = int(input("Quelle indice (position de l'élève dans le tableau) Voulez vous : "))
print("\nLa maison associé a l'élève est :", maison(tab, indice_eleve))
````

### Question 7

Suite aux ricanements de Severus, le professeur Remus Lupin vous propose de tester la fiabilité d'une méthode de répartition aléatoire des élèves. Vous ferez la moyenne sur 100 répartitions aléatoires pour avoir une valeur plus précise. Que peut-on en conclure ?

````python
#Bibliothèque importée
from random import*

#Initialisation des fonctions : 

def repartition_aleatoire(tab):
    
    """
    Cette fonction effectue une répartition aléatoire des maisons pour chaque élève.
    entrée 1 : tableau contenant le prénom et les notes des élèves, type list
    valeur retournée : renvoie un dictionnaire contenant le nom ainsi que la répartition aléatoire, type dict
    """
    
    nom_maison = ["Gryffondor","Serdaigle","Poufsouffle","Serpentard"] 
    d = {}
    d = repartition(tab)
    tab2 = list(d)
    i = 0
    
    while i < len(d):
        n = randint(0,3) #stocke un nombre aléatoire entre (0,1,2,3)
        d[tab2[i]] = nom_maison[n] #affectation d'une maison aléatoire a l'élève
        i = i + 1
        
    return d
    
def moyenne_100(d, d2):
    
    """
    Cette Fonction fait une moyenne de 100 répartition aléatoire.
    entrée 1 : dictionnaire contenant le prénom et les maison des élèves repartie aléatoirement , type dict
    entrée 2 : dictionnaire contenant le prénom et les maison des élèves fait par le choixpeau , type dict
    valeur Afficher : affiche le nombre d'erreurs et le pourcentage moyen de 100 repartition aléatoire.
    """
    
    i = 0
    somme = 0
    
    while i < 100:
        a = nb_erreurs(d, d2) # le nombre d'erreurs entre les dictionnaire d et d2
        somme = somme + a 
        i = i + 1
    moyenne = somme / i # moyenne des erreurs sur 100 répartition 
    print("le nombre d'erreur moyen entre la repartion aléatoire et celle du choixpeau est de : " ,moyenne)
    print("Ce qui représente ", round((moyenne*100) / len(d)),"% de faute en moyenne. \n ")
    
#Programme Principale :

d = repartition_aleatoire(tab) #le tableau avec la maison aleatoire
moyenne_100(d, d2)

print(
    "En conclusion, notre méthode de répartition des élèves est plus précise "
    "que la méthode aléatoire proposée par le professeur Remus Lupin. "
    "En effet, notre approche génère en moyenne 71 erreurs par rapport au choixpeau magique, "
    "alors que la répartition aléatoire entraîne environ 90 erreurs. "
    "Cela montre que notre méthode améliore la précision."
)
````
